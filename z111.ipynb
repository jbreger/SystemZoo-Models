Z111 Density-dependent growth (Michaelis Menten)

```python
#import libraries
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

```python
#Parameters
initial_stock = 0.02 #amount
max_growth_rate = 1  #1/day
harvest_rate = 0.5 # 1/day
half_saturation_constant = 1 #amount
```

```python
#Time
initial_time = 0
final_time = 50
time_step = 0.02
```

```python
t=np.arange(initial_time, final_time, time_step)
```

```python
def harvest(state, t, harvest_rate):
    return harvest_rate * state
```

```python
def increase(state, t, max_growth_rate, carrying_capacity):
    return max_growth_rate *state * (1-(state/half_saturation_constant + state))
```

```python
def ds(state, t, max_growth_rate, harvest_rate, carrying_capacity):
    return increase(state, t, max_growth_rate, carrying_capacity) - harvest(state, t, harvest_rate)
```

```python
state=odeint(ds, initial_stock, t, args=(max_growth_rate, harvest_rate, half_saturation_constant ,))
```

```python
#plot
plt.plot(t, state)

plt.title('Michael-Menten Kinectics')
plt.xlabel('Time (Year)')
plt.ylabel('Ammount/year')
```
