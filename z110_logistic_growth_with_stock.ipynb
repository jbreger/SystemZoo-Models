Z110 Logistic growth with stock-dependent harvest


```python
#import libraries
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

```python
#time
initial_time = 0
final_time = 20
time_step = 0.02

t=np.arange(initial_time, final_time, time_step)
```

```python
#Parameters
initial_stock = 0.01
max_growth_rate = 1
harvest_rate = 0.2
carrying_capacity = 1
```

```python
def harvest(state, t, harvest_rate):
    return harvest_rate * state
```

```python
def increase(state, t, max_growth_rate, carrying_capacity):
    return max_growth_rate *state * (1-(state/carrying_capacity))
```

```python
def ds(state, t, max_growth_rate, harvest_rate, carrying_capacity):
    return increase(state, t, max_growth_rate, carrying_capacity) - harvest(state, t, harvest_rate)
```

```python
state=odeint(ds, initial_stock, t, args=(max_growth_rate, harvest_rate, carrying_capacity,))
```

```python
#plot
plt.plot(t, state)

plt.title('Logistic growth with stock-dependent harvest')
plt.xlabel('Time (Year)')
plt.ylabel('Ammount/year')
```
