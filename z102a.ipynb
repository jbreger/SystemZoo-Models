```python
#import libraries
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

```python
#parameters
inflow_rate=10
initial_state=0
```

```python
#time_parameters
timezero=0
final_time=10
time_step=0.01
```

```python
def inflow (state, t, inflow_rate):
    return inflow_rate
```

```python
#create range
t=np.arange(timezero, final_time, time_step)
```

```python
state=odeint(inflow, initial_state, t, args=(inflow_rate,))
```

```python
#plot
plt.plot(t, state)
plt.xlabel('time')
plt.ylabel('in')
```
