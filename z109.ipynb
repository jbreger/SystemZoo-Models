Exponential Growth and Decay

```python
#import libraries
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

```python
#Parameters
capacity = 1
growth_rate = 1
initial_state=.01
```

```python
#Time parameters in years
Final_Time = 20
Initial_Time = 0
Time_Step = 0.05

t=np.arange(Initial_Time, Final_Time, Time_Step)
```

```python
def remaining_capacity_factor(state, t, capacity):
    return 1 -(state/capacity)
```

```python
def ds (state, t, capacity, growth_rate):
    return growth_rate * state * remaining_capacity_factor(state, t, capacity)
```

```python
state1=odeint(ds, initial_state, t, args=(capacity, growth_rate,))
```

```python
#plot
plt.plot(t, state1)

plt.title('Logistic Growth')
plt.xlabel('Time (Year)')
plt.ylabel('Ammount/year')
```

Z109B Logistic Growth at Constant Harvest

```python
#Parameters
initial_stock = 1
carrying_capacity = 1
max_growth_rate = 1
harvest_rate = .001
```

```python
#Time parameters in years
Final_Time = 20
Initial_Time = 0
Time_Step = 0.05

t=np.arange(Initial_Time, Final_Time, Time_Step)
```

```python
def harvest(state, t, harvest_rate):
    return harvest_rate
```

```python
def increase(state, t, max_growth_rate, carrying_capacity):
    return max_growth_rate * state * (1-(state/carrying_capacity))
```

```python
def ds2 (state, t, max_growth_rate, carrying_capacity, harvest_rate):
    return increase(state, t, max_growth_rate, carrying_capacity) - harvest(state, t, harvest_rate)
```

```python
state2=odeint(ds2, initial_stock, t, args=(max_growth_rate, carrying_capacity, harvest_rate))
```

```python
#plot
plt.plot(t, state2)

plt.title('Logistic Growth at Constant Harvest')
plt.xlabel('Time (Year)')
plt.ylabel('Ammount/year')
```
