Z107 Infection Dynamics

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

```python
##Parameters
Population = 4e007
Initial_infected = 10
Infections_per_contact=0.1
Frequency_of_Contact=0.2
beta = Infections_per_contact * Frequency_of_Contact
```

```python
def dI_dt(infected,t):
    Susceptible = Population - infected
    return beta * infected * (Susceptible/Population)
```

```python
t = np.linspace(0, 1500, 1500)
```

```python
infected = odeint(dI_dt, Initial_infected, t).flatten()
```

```python
S = Population -infected
```

```python
#  Plot 
plt.figure(figsize=(10, 6))

plt.plot(t, infected, label='Infected', lw=2)
plt.plot(t, S, label='Susceptible', lw=2)

plt.title("Simple SI model (no recovery)")
plt.xlabel("Time (days)")
plt.ylabel("Number of people")
plt.legend()
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```
