```python

#import libraries
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

```python
#make a step function
def step(t, time_on, value):
    if t < time_on:
        return 0
    else: 
        return value
    
```

```python
#Parameters
flow_rate_for_open_valve = 20
outflow_rate=0.5
state = 0
```

```python
#dynamics
def inflow( state, t, flow_rate_for_open_valve):
    return flow_rate_for_open_valve * step(t, 1, 1)
```

```python
def outflow(state, t, outflow_rate):
    return state * outflow_rate
```

```python

def dt (state, t, flow_rate_for_open_valve, outflow_rate):
    return inflow(state, t, flow_rate_for_open_valve) - outflow(state, t, outflow_rate)
```

```python
t=np.arange(0, 10, .01)
```

```python
state = odeint(dt, 0, t, args= (flow_rate_for_open_valve, outflow_rate))
```

```python
#plot
plt.plot(t, state)
plt.title('Tub with Leak')
plt.xlabel('Time (min)')
plt.ylabel('Liter')
```
