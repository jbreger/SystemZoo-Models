Exponential Growth and Decay

```python
#import libraries
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

```python
#Parameters
rate_of_change = -1
initial_state = 1
```

```python
#Time parameters
Final_Time = 10
Initial_Time = 0
Time_Step = 0.02

t=np.arange(Initial_Time, Final_Time, Time_Step)
```

```python
def ds1 (state, t, rate_of_change):
    return rate_of_change * state
```

```python
state1=odeint(ds1, initial_state, t, args=(rate_of_change,))
```

```python
def ds2 (state, t, rate_of_change):
    return -rate_of_change * state
```

```python
state2=odeint(ds2, initial_state, t, args=(rate_of_change,))
```

```python
#plot
plt.plot(t, state1)

plt.title('State and Change, Growth & Decay')
plt.xlabel('Time (Year)')
plt.ylabel('Ammount/year')
```

```python
#plot
plt.plot(t, state2)

plt.title('State and Change, Growth & Decay')
plt.xlabel('Time (Year)')
plt.ylabel('Ammount/year')
```
